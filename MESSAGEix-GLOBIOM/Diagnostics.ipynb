# Calculate VRE penetration levels

```python
tech_colors = {'solar':"#f9d002",
               'wind':"#235ebc"}
```

```python
VRE_penetration, VRE_cap, VRE_act = calculate_VRE_penetration(base_act,base_cap,EU_regions)

fig,ax = plt.subplots()
VRE_penetration.plot.bar(ax=ax, 
                         stacked=True,
                         color = [tech_colors[i] for i in VRE_penetration.columns])
ax.set_ylabel("VRE penetration level (% of electricity supply)")
ax.set_ylim(0,100)
fig.savefig("figures/base_VRE_share.png",bbox_inches="tight",dpi=300)

fig,ax = plt.subplots()
VRE_cap.plot.bar(ax=ax, 
                 stacked=False,
                 color = [tech_colors[i] for i in VRE_cap.columns])
ax.set_ylabel("VRE capacity (GW)")
fig.savefig("figures/base_VRE_cap.png",bbox_inches="tight",dpi=300)

fig,ax = plt.subplots()
VRE_act.plot.bar(ax=ax, 
                 stacked=False,
                 color = [tech_colors[i] for i in VRE_act.columns])
ax.set_ylabel("VRE activity (GWa)")
ax.set_ylim(0,650)
fig.savefig("figures/base_VRE_act.png",bbox_inches="tight",dpi=300)
```
